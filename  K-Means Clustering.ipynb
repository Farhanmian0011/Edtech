```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
# Create dummy data: Customer spending habits
np.random.seed(42)  # For consistent results
data = {
    'Annual_Income': [20, 25, 30, 35, 40, 60, 65, 70, 75, 80, 90, 95, 100, 105, 110],
    'Spending_Score': [10, 15, 20, 25, 30, 70, 75, 80, 85, 90, 40, 45, 50, 55, 60]
}
df = pd.DataFrame(data)
print(df)
```

```
plt.scatter(df['Annual_Income'], df['Spending_Score'])
plt.xlabel('Annual Income (in thousands)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Data')
plt.show()
```

```
# Prepare data
X = df[['Annual_Income', 'Spending_Score']]

# Create K-Means model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster labels for each point
df['Cluster'] = kmeans.labels_
```

```
# Plot different clusters in different colors
plt.scatter(df['Annual_Income'], df['Spending_Score'], c=df['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='x', s=200, label='Centroids')
plt.xlabel('Annual Income (in thousands)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering Results')
plt.legend()
plt.show()
```

```
print("Data with Cluster Labels:")
print(df)
print(f"\nCluster Centers:")
print(kmeans.cluster_centers_)
```

## Key Observations
- K-Means clustering grouped customers into 3 distinct clusters.
- Each cluster represents different spending patterns based on income and spending score.
- The red X marks show the center points of each cluster.
- Cluster 0: Low income, low spending
- Cluster 1: High income, high spending  
- Cluster 2: Medium income, medium spending
- This helps businesses understand different customer segments.
